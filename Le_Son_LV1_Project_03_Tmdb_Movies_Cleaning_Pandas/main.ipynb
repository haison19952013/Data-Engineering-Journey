```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
import numpy as np
```

```python
df = pd.read_csv('tmdb-movies.csv')
df.head()
```

```python
def text_to_datetime(text):
  text = text.split('/')
  if int(text[2]) < 30:
    text[2] = '20' + text[2]
  else:
      text[2] = '19' + text[2]
  return f'{int(text[0]):02d}/{int(text[1]):02d}/{int(text[2]):04d}'

def cast_genres_to_list(text):
  text = text.split('|')
  return text

def wrangle(df):
  # Drop duplicate and missing values
  df = df.drop_duplicates()
  df = df.dropna(subset=['id'])
  # Convert genres to list
  df['genres'].fillna('', inplace=True)
  df['genres'] = df['genres'].apply(cast_genres_to_list)
  # Convert cast to list
  df['cast'].fillna('', inplace=True)
  df['cast'] = df['cast'].apply(cast_genres_to_list)
  # Convert release_date to datetime
  df['release_date'] = df['release_date'].apply(text_to_datetime)
  df['release_date'] = pd.to_datetime(df['release_date'], format=r'%m/%d/%Y')
  # Add profit columns
  df['profit'] = df['revenue_adj'] - df['budget_adj']
  # Add ROI columns
  df['ROI'] = df['profit'] / df['budget_adj'] * 100
  return df
df_wrangle = wrangle(df)
df_wrangle.info()
```

1. Sắp xếp các bộ phim theo ngày phát hành giảm dần rồi lưu ra một file mới

```python
df_wrangle.sort_values(ascending=False, by='release_date').to_csv('tmdb-movies-release-date-sort.csv', index=False)
```

2. Lọc ra các bộ phim có đánh giá trung bình trên 7.5 rồi lưu ra một file mới


```python
df_wrangle[df_wrangle['vote_average'] > 7.5].to_csv('tmdb-movies-vote-average-gt75.csv', index=False)
```

3. Tìm ra phim nào có doanh thu cao nhất và doanh thu thấp nhất


```python
min_revenue = df_wrangle['revenue_adj'].min()
max_revenue = df_wrangle['revenue_adj'].max()
df_wrangle[df_wrangle['revenue_adj'] == min_revenue].to_csv('tmdb-movies-revenue-adj-min.csv', index=False)
df_wrangle[df_wrangle['revenue_adj'] == max_revenue].to_csv('tmdb-movies-revenue-adj-max.csv', index=False)
```

```python
df_wrangle.sort_values(ascending=False, by='profit').head(n = 10).to_csv('tmdb-movies-top10-profit.csv', index=False)
```

6. Đạo diễn nào có nhiều bộ phim nhất và diễn viên nào đóng nhiều phim nhất


```python
df_wrangle.groupby('director').size().sort_values(ascending=False).rename('count').to_frame().head(n = 1).to_csv('tmdb-movies-director-most.csv')
```

```python
df_cast = df_wrangle.explode('cast').groupby('cast').size().sort_values(ascending=False).rename('count').to_frame().reset_index()
df_cast[df_cast['cast'] != ''].head(n = 1).to_csv('tmdb-movies-cast-most.csv', index=False)
```

7. Thống kê số lượng phim theo các thể loại. Ví dụ có bao nhiêu phim thuộc thể loại Action, bao nhiêu thuộc thể loại Family, ….


```python
df_genres = df_wrangle.explode('genres').groupby('genres').size().sort_values(ascending=False).rename('count').to_frame().reset_index()
df_genres[df_genres['genres'] != ''].to_csv('tmdb-movies-genres-count.csv', index=False)
```

```python
fig, ax = plt.subplots(figsize=(6, 4))
famous_director_MedROI = df_wrangle[df_wrangle['director'] == 'Woody Allen']['ROI'].replace([np.inf, -np.inf], np.nan).dropna().median()
df_wrangle[df_wrangle['budget_adj'] != 0].groupby('release_year')['ROI'].median().plot(ax = ax)
ax.axhline(y=famous_director_MedROI, color='r', linestyle='--', label = 'Woody Allen')
ax.legend()
ax.set_xlabel('Year')
ax.set_ylabel('Median of ROI (%)')
ax.set_title('Median of ROI by Year');
```

```python
fix, ax = plt.subplots(figsize=(6, 6))
df_wrangle.explode('genres').groupby('genres')['ROI'].median().replace([np.inf, -np.inf], np.nan).dropna().sort_values(ascending=True).plot(kind='barh', ax = ax)
ax.set_xlabel('Median of ROI (%)')
ax.set_ylabel('Genres')
ax.set_title('Median of ROI by Genres');
```

```python
fix, ax = plt.subplots(figsize=(6, 6))
df_wrangle.explode('cast').groupby('cast')['ROI'].median().replace([np.inf, -np.inf], np.nan).dropna().sort_values(ascending=False).head(n = 20).sort_values(ascending=True).plot(kind='barh', ax = ax)
ax.set_xlabel('Median of ROI (%)')
ax.set_ylabel('Cast')
ax.set_title('Median of ROI by Cast');
```

```python
fig, ax = plt.subplots(figsize=(6, 4))
peason_corr = df_wrangle[['budget_adj', 'revenue_adj']].corr().iloc[0, 1]
sns.regplot(x='budget_adj', y='revenue_adj', data=df_wrangle, ax = ax)
ax.set_xlabel('Budget (Adjusted)')
ax.set_ylabel('Revenue (Adjusted)')
ax.set_title('Revenue vs Budget');
ax.scatter(df_wrangle['budget_adj'], df_wrangle['revenue_adj'], alpha=0.5)
ax.set_title(f'Revenue vs Budget (Peason Correlation: {peason_corr:.2f})');
```

```python
fig, ax = plt.subplots(figsize=(6, 4))
df_right_ROI = df_wrangle.copy()
df_right_ROI[['ROI', 'popularity']] = df_right_ROI[['ROI', 'popularity']].replace([np.inf, -np.inf], np.nan)
df_right_ROI = df_right_ROI.dropna(subset=['ROI', 'popularity'])
peason_corr = df_right_ROI[['ROI', 'popularity']].corr().iloc[0, 1]
sns.regplot(x='popularity', y='ROI', data=df_right_ROI, ax=ax)
ax.set_yscale('log')
ax.set_ylabel('ROI (%)')
ax.set_xlabel('Popularity ')
ax.set_title(f'Popularity vs ROI (Peason Correlation: {peason_corr:.2f})');
```
